```python
import itertools
import pandas as pd
```

In this example there are attributes of a computer processing system being observed: `CH1`, `CH2`, and `CPU`, each of which can either be at high (`H`) or low (`L`) processing power. Assume that over a period of monitoring, the following combinations of `CH1`, `CH2`, `CPU` are observed:

| `CH1` | `CH2` | `CPU` |
|--|--|--|
| `L` | `L` | `L` |
| `L` | `H` | `H` |
| `H` | `L` | `H` |
| `H` | `H` | `L` |


```python
perm = [list(i) for i in list(itertools.product([0,1], repeat=8))]
```

```python
print("First 5 permutations of CH1-CH2-CPU")
perm[:5]
```

```python
# Create a DataFrame with column labels from permutation dataset
df = pd.DataFrame(perm)
# df["perm"].to_list()
df.columns = ['LLL', 'LLH', 'LHL', 'LHH', 'HLL', 'HLH', 'HHL', 'HHH']
df.head()
```

```python
# Pairwise observations: CH1-CH2
df['CH1_CH2_LL'] = ''
df['CH1_CH2_LH'] = ''
df['CH1_CH2_HL'] = ''
df['CH1_CH2_HH'] = ''

# Pairwise observations: CH1-CPU
df['CH1_CPU_LL'] = ''
df['CH1_CPU_LH'] = ''
df['CH1_CPU_HL'] = ''
df['CH1_CPU_HH'] = ''

# Pairwise observations: CH2-CPU
df['CH2_CPU_LL'] = ''
df['CH2_CPU_LH'] = ''
df['CH2_CPU_HL'] = ''
df['CH2_CPU_HH'] = ''
```

```python
df.head()
```

For each permutation of the CH1-CH2-CPU ternary relation, match the CH1 and CH2 observation to a pattern match for the binary CH1-CH2 relation.

```python
# CH1-CH2 observed Low, Low
for index, row in df.iterrows():
    if row["LLL"] == 1 or row["LLH"] == 1:
        df.at[index, "CH1_CH2_LL"] = 1

# CH1-CH2 observed Low, High
for index, row in df.iterrows():
    if row["LHL"] == 1 or row["LHH"] == 1:
        df.at[index, "CH1_CH2_LH"] = 1

# CH1-CH2 observed High, Low
for index, row in df.iterrows():
    if row["HLL"] == 1 or row["HLH"] == 1:
        df.at[index, "CH1_CH2_HL"] = 1

# CH1-CH2 observed High, High
for index, row in df.iterrows():
    if row["HHL"] == 1 or row["HHH"] == 1:
        df.at[index, "CH1_CH2_HH"] = 1
```

```python
# CH1-CPU observed Low, Low
for index, row in df.iterrows():
    if row["LLL"] == 1 or row["LHL"] == 1:
        df.at[index, "CH1_CPU_LL"] = 1

# CH1-CPU observed Low, High
for index, row in df.iterrows():
    if row["LLH"] == 1 or row["LHH"] == 1:
        df.at[index, "CH1_CPU_LH"] = 1

# CH1-CPU observed High, Low
for index, row in df.iterrows():
    if row["HLL"] == 1 or row["HHL"] == 1:
        df.at[index, "CH1_CPU_HL"] = 1

# CH1-CPU observed High, High
for index, row in df.iterrows():
    if row["HLH"] == 1 or row["HHH"] == 1:
        df.at[index, "CH1_CPU_HH"] = 1
```

```python
# CH2-CPU observed Low, Low
for index, row in df.iterrows():
    if row["LLL"] == 1 or row["HLL"] == 1:
        df.at[index, "CH2_CPU_LL"] = 1

# CH2-CPU observed Low, High
for index, row in df.iterrows():
    if row["LLH"] == 1 or row["HLH"] == 1:
        df.at[index, "CH2_CPU_LH"] = 1

# CH2-CPU observed High, Low
for index, row in df.iterrows():
    if row["HHL"] == 1 or row["LHL"] == 1:
        df.at[index, "CH2_CPU_HL"] = 1

# CH2-CPU observed High, High
for index, row in df.iterrows():
    if row["LHH"] == 1 or row["HHH"] == 1:
        df.at[index, "CH2_CPU_HH"] = 1
```

```python
# Replace blank values with 0
df.replace('', 0, inplace=True)
```

```python
binary_relation_cols = list(df.columns)[8:]
df['total'] = df[binary_relation_cols].sum(axis=1)
```
