# Symbolic Computing

```python
import sympy as sy
sy.init_printing()
```

```python
x, y, z, a, b, c = sy.symbols('x y z a b c')
```

```python
x**2
```

```python
sol = sy.solve(a*x**2 + b*x + c, x)
sol
```

```python
sol[0].subs([(x, 1), (a, 2), (b, 2), (c, 1)])
```

```python
# Plotting function
f = (x**2 + y - 11)**2 + (x + y**2 - 7)**2
sy.plot(f.subs([(y, 6)]), (x, -6, 6))
```

```python
# Plotting range of functions
f = (x**2 + y - 11)**2 + (x + y**2 - 7)**2

for i, yval in enumerate(range(-6, 7)):
    if i == 0:
        p = sy.plot(f.subs([(y, yval)]), (x, -6, 6))
    else:
        p.extend(sy.plot(f.subs([(y, yval)]), (x, -6, 6)))

p.show()
    
```
